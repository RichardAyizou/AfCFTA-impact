# PIB des pays de l'UEMOA

```python
import wbgapi as wb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
waemu = ['BEN', 'BFA', 'CIV', 'GNB', 'MLI', 'NER', 'SEN','TGO']
```

```python
# GDP (constant 2015 US$)
wb.data.DataFrame('NY.GDP.MKTP.KD', waemu, time=range(1970,2019), labels=True)
```

```python
wb.data.DataFrame('NY.GDP.MKTP.KD', waemu, time=range(1970,2019), index='time').plot()
plt.show()
```

```python
# GDP growth (annual %)

wb.data.DataFrame('NY.GDP.MKTP.KD.ZG', waemu, time=range(1970,2019), index='time').plot()
plt.show()
```

```python
# GDP per capita (constant 2015 US$)
wb.data.DataFrame('NY.GDP.PCAP.KD', waemu, time=range(1970,2019), index='time').plot()
plt.show()
```

```python
# Population
wb.data.DataFrame('SP.POP.TOTL', waemu, time=range(1970,2019), index='time').plot()
plt.show()
```
